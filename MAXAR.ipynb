Access MAXAR image collections

```python
leafmap.maxar_collections()
```

Number of avaiable collections

```python
len(leafmap.maxar_collections())
```

Couting the number of images in each collection

```python
collection_counts = {}
for collection in leafmap.maxar_collections():
  collection_counts[collection] = len(leafmap.maxar_search(collection))

for collection, count in collection_counts.items():
  print(f"Collection: {collection}, Image Count: {count}")
```

Couting the total number of images

```python
total_images = sum(collection_counts.values())
print(f"Total number of images: {total_images}")
```

Select the Brazil-Flooding-May24 collection and sort by ID


```python
collections = leafmap.maxar_child_collections('Brazil-Flooding-May24')
print(f"The number of collections: {len(collections)}")
collections.sort()

# Create an empty vector to store the collection codes
collection_codes = []

# Iterate through the collections and append the codes to the vector
for collection in collections:
  collection_codes.append(collection)

# Print the vector of collection codes
print(collection_codes)
```

```python
gdf = leafmap.maxar_items(
    collection_id='Brazil-Flooding-May24',
    child_id= collection_codes[1],
    return_gdf=True,
    assets=['visual'])
gdf
```

```python
number_images = len(gdf)
```

Plotting the footprints on the map

```python
m = leafmap.Map()
m.add_gdf(gdf, layer_name="Footprints", zoom_to_layer=True, info_mode="on_click")
m
```

Generating the COG URL for all items in the 10300100890F4800 collection

```python
images = gdf['visual'].tolist()
images[:number_images]
```

Downloading COGs to a local directory

```python
leafmap.maxar_download(images[:number_images])
```
